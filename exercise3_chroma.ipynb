```python
import chromadb
from sentence_transformers import SentenceTransformer
print(f"chromaDB version: {chromadb.__version__}")
print("chromaDB imported successfully!")
# print("library successfully imported")
```

```python
# Load embedding model
model = SentenceTransformer("all-MiniLM-L6-v2")
print("model loaded successfully")
```

```python
# Create Chroma client
# client = chromadb.Client()


# Create Chroma client (persistent storage)

client = chromadb.PersistentClient(path="./chroma_db")

# create or get collection
collection = client.get_or_create_collection(name="my_documents_management", 
metadata={"description": "Document Management collection"}
)

# IDs for each document
ids = [f"doc_{i}" for i in range(len(documents))]

print(f"collection created: {collection.name}")
print(f"Current count: {collection.count()} documents")
print(f" Data persisted to: ./chroma_db/")
```

```python
texts = [doc["text"] for doc in documents]
metadatas = [doc["metadata"] for doc in documents]
ids = [f"doc_{i}" for i in range(len(texts))]

# Generate embeddings
embeddings = model.encode(texts).tolist()

# Add documents to chroma

collection.add(
    documents=texts,
    metadatas=metadatas,
    ids=ids,
    embeddings=embeddings
)

print(f"Added {len(documents)} documents to collection")
print(f"Total documents: {collection.count()}")
```

```python
# Query the collection
# by category
query_text = "Advances in robotics reshape manufacturing..."
results = collection.query(
    query_texts=[query_text],
    n_results=2,
    where={"category": "tech"}
)

print(results["documents"])
```

```python
# Query the collection
# by author

results = collection.query(
    query_texts=[query_text],
    n_results=2,
    where={"author": "Henry Adams"}
)

print(results["documents"])
```

```python
# Query the collection
# by author

results = collection.query(
    query_texts=[query_text],
    n_results=2,
    where={"year": {"$gte": 2024}}
)
# 2024-03-03
print(results["documents"])
```
